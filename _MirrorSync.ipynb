Import libraries

```python
import os
import shutil
```

Explore folder

```python
folder1 = input("Enter the first folder: ")
folder2 = input("Enter the second folder: ")
```

Get files info for folder1

```python
file_info = []
counter = 0
for root, _, files in os.walk(folder1):
    for file in files:
      counter += 1
      if counter % 100 == 0:
        print("scanned: " + str(counter))
      # Check for hidden files (indented under the loop)
      file_path = os.path.join(root, file)
    
      # Handle potential errors during file size retrieval
      try:
          file_size = os.path.getsize(file_path)
          if file_size > 0:  # Only process non-empty files (0kb)
              file_info.append((file, root, file_size))
      except FileNotFoundError:
          # Handle potential file not found errors (like $RECYCLE.BIN)
          pass  # Ignore the error and continue processing other files
```

Get files info for folder2

```python
file_info_2 = []
counter = 0
for root, _, files in os.walk(folder2):
    for file in files:
      counter += 1
      if counter % 100 == 0:
        print("scanned: " + str(counter))
      # Check for hidden files (indented under the loop)
      file_path = os.path.join(root, file)
    
      # Handle potential errors during file size retrieval
      try:
          file_size = os.path.getsize(file_path)
          if file_size > 0:  # Only process non-empty files (0kb)
              file_info_2.append((file, root, file_size))
      except FileNotFoundError:
          # Handle potential file not found errors (like $RECYCLE.BIN)
          pass  # Ignore the error and continue processing other files
```

Change of variable and print

```python
folder1_info = file_info
folder2_info = file_info_2
```

Print sublist.

```python
sublist1 = []
sublist2 = []

len_folder1_info=len(folder1_info)
len_folder2_info=len(folder2_info)


counter = 0
for file, root, size in folder1_info:
    counter += 1
    if counter % 100 == 0:
        print(counter / len(folder1_info))  # Option for progress tracking
    # Check for (file, size) only in folder2_info
    if (file, size) not in [(f, s) for f, _, s in folder2_info]:
        sublist1.append((file, root, size))  # Append the full tuple
        

counter=0
for file, root, size in folder2_info:
    counter += 1
    if counter % 100 == 0:
        print(counter / len(folder1_info))  # Option for progress tracking
    # Check for (file, size) only in folder2_info
    if (file, size) not in [(f, s) for f, _, s in folder1_info]:
        sublist2.append((file, root, size))  # Append the full tuple
        
```

Transfer.

```python
for file, root, _ in sublist1:
    source_path = os.path.join(root, file)
    dest_path = os.path.join(folder2, "transfer", os.path.relpath(source_path, folder1))
    os.makedirs(os.path.dirname(dest_path), exist_ok=True)  # Create intermediate directories
    shutil.copy2(source_path, dest_path)  # Copy file with metadata
    print(dest_path)

for file, root, _ in sublist2:
    source_path = os.path.join(root, file)
    dest_path = os.path.join(folder1, "transfer", os.path.relpath(source_path, folder2))
    os.makedirs(os.path.dirname(dest_path), exist_ok=True)  # Create intermediate directories
    shutil.copy2(source_path, dest_path)
    print(dest_path)# Copy file with metadata
```
